```python
# Case Folding: Lowercase
# Merubah format teks menjadi format huruf kecil semua

kalimat = "Ini ADalah cOnToh daRI CasE fOlDINg LowER caSE!"
lower_case = kalimat.lower()
print(lower_case)
```

```python
# Case Folding: Removing Number
# Menghapus karakter angka

import re

kalimat = "Sebanyak 21 siswa berprestasi memperoleh beasiswa pendidikan tinggi."
hasil = re.sub(r"\d+", "", kalimat)
print(hasil)
```

```python
# Case Folding: Removing Punctuation
# Menghapus karakter tanda baca

import string

kalimat = "Ini adalah contoh teks dengan tanda baca: !@#$%^&*()."
# Use the built-in str.maketrans to remove punctuation
hasil = kalimat.translate(str.maketrans("", "", string.punctuation))
print(hasil)
```

```python
# Case Folding: Removing Whitespace
# Menghapus karakter kosong

kalimat = "\t     Ini adalah contoh teks dengan karakter kosong.  \t \n"
print(kalimat) #output kalimat asli

hasil = kalimat.strip()
print(hasil)
```

```python
# Seperating Sentences with Split () Method
# Fungsi split() memisahkan string ke dalam list dengan spasi sebagai pemisah jika tidak ditentukan pemisahnya.

kalimat = "Ini adalah contoh teks-teks yang akan dipecah dengan spasi sebagai pemisah-nya."
hasil = kalimat.split()
print(hasil)
```

```python
# Tokenizing with Case Folding
# Menggabungkan teknik Case Foling dengan Tokenizing.

import string
import nltk
from nltk.tokenize import word_tokenize

kalimat = "Ini ADalah cOnToh daRI CasE fOlDINg DeNg.an TokeN!"
# Use the built-in str.maketrans to remove punctuation
kalimat = kalimat.translate(str.maketrans('','',string.punctuation)).lower()

tokens = nltk.tokenize.word_tokenize(kalimat)
print(tokens)
```

```python
# Frequency Distribution
# Menghitung frekuensi kemunculan setiap tokens(kata) dalam teks.

import string
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

kalimat = "Ibu seringkali membeli kebutuhan rumah di Pasar, padahal di zaman sekarang, Ibu bisa membeli kebutuhan rumah dengan belanja online."
# Use the built-in str.maketrans to remove punctuation
kalimat = kalimat.translate(str.maketrans('','',string.punctuation)).lower()

tokens = nltk.tokenize.word_tokenize(kalimat)
freq_dist = nltk.FreqDist(tokens) #menghitung kemunculannya

print(freq_dist.most_common())
```

```python
# Frequency Distribution Visualization with Matplotlib
# Untuk menggambarkan frekuensi kemunculan setiap tokens dapat menggunakan library matplotlib pada Python.

import matplotlib.pyplot as plt

freq_dist.plot(30, cumulative=False)

plt.show()
```

```python
# Filtering using NLTK
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import string # Import string module

nltk.download('stopwords')

kalimat = "Menurut hakim, proses penyidikan yang dijalankan Kejaksaan Agung sudah sesuai dengan prosedur hukum acara yang berlaku. Kejaksaan Agung memulai proses hukum dengan melakukan penyelidikan pada 20 Mei 2025. Selanjutnya diterbitkan Surat Perintah Penyidikan (Sprindik) pada 11 Juni 2025."
# Use the built-in str.maketrans to remove punctuation
kalimat = kalimat.translate(str.maketrans('','',string.punctuation)).lower() #lower case dan hapus tanda baca

tokens = word_tokenize(kalimat) #tokenize per word
listStopword = set(stopwords.words('indonesian')) #daftar stopword dalam bahasa indonesia

removed = []
for t in tokens:
    if t not in listStopword:
        removed.append(t) #menghapus stopword dalam tokenize

print(removed)
```

```python
# Stemming : Porter Stemming Algorithm using NLTK

from nltk.stem import PorterStemmer
ps = PorterStemmer()

kata = ["program", "programs", "programer", "programing", "programers"]

for k in kata:
    print(k, " : ", ps.stem(k))
```
